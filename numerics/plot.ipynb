```julia
using Plots
```

```julia
function getConvexPlot(x::ExpSum)
    cs = Array{Float64,1}[]

    for i in 1:length(x.ExpList)
        term = x.ExpList[i]
        re = real(term)
        im = -1 * imag(term)
        push!(cs, [re, im])
    end

    hull = convex_hull(cs)
    
    p = plot([Singleton(ci) for ci in cs])
    plot!(p, VPolygon(hull), alpha=0.2, aspect_ratio=:equal)

end
```

```julia
a = ExpSum(Complex{Float64}[1.0 + 0.0im, 2.0 + 1.0im, 3.0 + 0.0im, 2.0 + 0.0im], Complex{Float64}[2.45 + 0.0im, 5.0 + 1.0im, 7.01 + 0.0im, 7.01 + 0.0im])
```

```julia
b = ExpSum(Complex{Float64}[3im, -3im], Complex{Float64}[0.5,0.5])
```

```julia
getConvexPlot(a)
```

```julia
getConvexPlot(b)
```
